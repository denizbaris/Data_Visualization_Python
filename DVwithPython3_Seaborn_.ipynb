```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
import warnings
warnings.filterwarnings('ignore')
```

```python
langs=["C","C++","Java","Python","PHP"]
students=[23,17,35,29,12]
fig,ax=plt.subplots()
ax.bar(langs,students)
for i in range(len(langs)):
    plt.text(langs[i],students[i],str(students[i]),ha="center",va="bottom")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

##Distribution Plots: histplot()

```python
sns.histplot(data = tips, x = 'total_bill', bins = 10, kde = True)   #bagzi dagilimlarda dagilimi daha net gormek icin bins'i arttir
```

##Adding in a grid and styles

```python
sns.set(style='darkgrid')
sns.histplot(data = tips, x = 'total_bill',bins=20)
```

```python
sns.set(style='darkgrid', rc={"grid.color": ".6", "grid.linestyle": ":"})
sns.histplot(data = tips, x = 'total_bill',bins=20);
```

```python
sns.displot(data = tips, x = 'total_bill',bins=20,kde=False,
            color='red',edgecolor='black',lw=4,ls='--')
```

##Cut Off KDE

```python
sns.kdeplot(tips.total_bill, clip = [10,20]);
```

##Bandwith

```python
sns.kdeplot(x = 'total_bill', data = tips, bw_adjust = 0.2, shade= True, color = "red");  #shade= True, color = "red" alani doldurur rengini secer
```
