```python
# Basic Libraries
import pandas as pd
import numpy as np
# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
titanic  = sns.load_dataset("titanic")
titanic
```

```python
sns.catplot(x = "pclass", y = "survived", data = titanic, kind = "bar", 
            estimator=np.sum, hue = "sex", col = "alone", row = "embarked" );
```

```python
sns.catplot(x = "pclass", y = "survived", data = titanic, kind = "bar", 
            estimator=np.sum, col = "sex",);
```

```python
tips = sns.load_dataset("tips")
tips.head(2)
```

```python
sns.catplot(x ="day", y="total_bill", data= tips, kind="box");  #farkli subplotlari ayni sekil icinde kullanicaksan catplotla kolayca yapabilirsin
```

```python
sns.catplot(x ="day", y="total_bill", data= tips, kind="violin", hue = "smoker");
```

```python
sns.catplot(x ="pclass", y="survived", data= titanic, kind="bar", 
            estimator= np.count_nonzero, col = "sex");
```

##Jointplot

```python
sns.jointplot(x = 'total_bill', y = "tip", data=tips, kind = "hex");
```

```python
sns.jointplot(x = 'total_bill', y = "tip", data=tips, kind = "reg");
```

##Pairplot

```python
sns.pairplot(tips, hue="sex", palette= "rocket_r", corner=True);
```

##Pairgrid

```python
g = sns.PairGrid(tips)
g = g.map(sns.scatterplot)
```

```python
g = sns.PairGrid(tips, hue="sex", palette="viridis")
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.distplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();
```

##FacetGrid

```python
g = sns.FacetGrid( data = tips, col = "time", row = "smoker")
g = g.map(plt.hist, "total_bill")
```

##Heatmap

```python
sns.heatmap(data=tips.corr(), linecolor='white',cmap='viridis', annot=True);
```
