```python
# Basic Libraries
import pandas as pd
import numpy as np
# Libraries for visualizaton
import matplotlib.pyplot as plt
import seaborn as sns
#print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset ("tips")
tips.head()
```

```python
tips.describe()
```

##Countplot

```python
tips.day.value_counts()
```

```python
sns.countplot(x= "day", data = tips);   #y ekseninde numeric columnlari count eder
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(x='day', data=tips)
for p in ax.patches:           #sutunlarin ustune numaralari yaz
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+0.5));
    # matplotlib.pyplot.annotate(text, xy, *args, **kwargs)
    # [https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html]
```

```python
print(ax.patches)
```

```python
for p in ax.patches:
  print(p)
```

```python
sns.countplot(x = "day", data = tips, hue = "sex");
```

```python
ax = sns.countplot(x='day', data=tips, hue = "sex")
for p in ax.patches:           #sutunlarin ustune numaralari yaz
    ax.annotate((p.get_height()), (p.get_x()+0.1, p.get_height()+0.5))
    ax.text(p.get_x()+0.1, p.get_height()-3,str(round(p.get_height()/tips.day.count(),2)) )     # yüzdelik ekledik
```

##BarPlot
Countplot da x'e kategorik degeri veryorsun ve onun count'unu gosteriyor


---

barplot'da x ve y veriyorsun hem kategorik hem numeric

```python
tips.describe()
```

```python
sns.barplot(x='sex', y = 'total_bill', data= tips);
```

```python
sns.barplot(x='day', y="total_bill",data=tips, hue='sex', ci='sd');  #conficance intervallari standart deviation
```

```python
tips.groupby(['day', 'sex']).sum()
```

estimator param (np.mean, np.median, np.sum, np.max, np.min, np.count_nonzero)

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
sns.barplot(x="day", y="total_bill", data=tips, ax = ax[0])
sns.countplot(x="day", data=tips, ax = ax[1])
plt.tight_layout()
```

##BoxPLot

```python
sns.boxplot(x = 'day' , y = 'total_bill', data= tips);
```

```python
sns.boxplot(x = 'day' , y = 'total_bill', data= tips, hue='sex');  #legend outlierlarin gorunumunu engelliyor
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0)
```

##Optional boxplot examples

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
sns.boxplot(x = "parental level of education", y = "math score", data = df)
ploe = df["parental level of education"].value_counts()
ploe_pos = range(len(ploe.index))
plt.xticks(ploe_pos, ploe.index, rotation=45);
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='parental level of education',y='math score',data=df,hue='gender')
# Optional move the legend outside
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.xticks(rotation = 'vertical');
```

##Violinplot

```python
sns.violinplot(x ="day", y = "total_bill", data = tips, hue="sex");
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x ='day', y = 'total_bill', data=tips, hue='sex', 
               inner= "quartile")  #q1 q2 ve q3 degerlerini inner ile uyguladik
```

```python
plt.figure(figsize=(12,6))
sns.violinplot(x ="day", y = "total_bill", data = tips, hue="sex", bw= 0.2);  
#bandwith
```

##Swarmplot

```python
sns.swarmplot(x="day",y="total_bill",data=tips, hue="sex");
```

##Line plot

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
sns.lineplot(x = 'year', y = 'passengers', data = flights)
```

```python
f_sum = flights.groupby(["year", "month"]).sum()
```

```python
sns.lineplot(x= f_sum.passengers, y= f_sum.reset_index().index)
```

```python
flights_wide = flights.pivot(index = "year", columns = "month", values = "passengers") #pivot ettik yani transform
flights_wide
```

```python
sns.lineplot(data = flights_wide)
```
